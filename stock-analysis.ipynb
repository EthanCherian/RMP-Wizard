# Stock Sentiment Analysis
Given a set of news headlines for a particular company, can you predict whether their stock price will rise or fall?

```python
import pandas as pd

df = pd.read_csv("Data.csv", encoding="ISO-8859-1")
df.head(3)           # label indicates whether stock price rose or fell (1: rose, 0: fell or stayed same)
```

```python
# Train Test Split
train = df[df["Date"] < "20150101"]
test = df[df["Date"] > "20141231"]
```

```python
# Removing Punctuations
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)

# Renaming Columns
list1 = [i for i in range(25)]
new_Index = [str(i) for i in list1]
data.columns = new_Index
data.head(3)
```

```python
# Converting headlines to lowercase
for index in new_Index:
    data[index] = data[index].str.lower()
data.head(1)
```

```python
# Combine entire row into single paragraph to be considered using term frequency
headlines = []
for row in range(0, len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[row, 0:25]))
headlines[0]
```

```python
# Create Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
train_ds = cv.fit_transform(headlines)
```

```python
# Create Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
randomClassifier = RandomForestClassifier(n_estimators=200, criterion="entropy")
randomClassifier.fit(train_ds, train["Label"])
```

```python
# Predict for Test Dataset
test_transform = []
for row in range(0, len(test.index)):
    test_transform.append(" ".join(str(x) for x in test.iloc[row, 2:27]))
test_ds = cv.transform(test_transform)
predictions = randomClassifier.predict(test_ds)
```

```python
# Check Accuracy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

matrix = confusion_matrix(test["Label"], predictions)
print(matrix)
print()

score = accuracy_score(test["Label"], predictions)
print("Score: " + str(score) + "\n")

report = classification_report(test["Label"], predictions)
print(report)
```

Matrix reads as:

    1. Top left: 1's correctly classified as 1's

    2. Top right: 1's incorrectly classified as 0's

    3. Bottom left: 0's incorrectly classified as 1's

    4. Bottom right: 0's correctly classified as 0's
